# Euro 12

## Import data from datasets/euro12.csv

```python
FILEPATH = '../data/euro12.csv'
```

##  Calculate the mean Yellow Cards given per Team

```python
euro12_data = pd.read_csv(FILEPATH, sep=',')
print("Euro 2012 dataset imported from euro12.csv:\n", euro12_data)
```

```python
mean_yellow_cards = euro12_data['Yellow Cards'].mean()
print(f"Mean number of yellow cards per team: {mean_yellow_cards}")
```

## Which teams scored more than 6 goals?

```python
for row in euro12_data['Goals']:
    if row > 6:
        print(f"Teams that scored more than 6 goals: {euro12_data.loc[euro12_data['Goals'] == row, 'Team'].iloc[0]}") 
```

## Select the teams that start with G

```python
for row in euro12_data['Team']:
    if row.startswith('G'):
        print(f"Teams that start with 'G': {euro12_data.loc[euro12_data['Team'] == row, 'Team'].iloc[0]}")
```

## Present only the Shooting Accuracy from England, Italy and Russia

```python
for row in euro12_data['Team']:
    if row in ['England', 'Italy', 'Russia']:
        print(f'Shooting accuracy of England, Italy, and Russia: {euro12_data.loc[euro12_data["Team"] == row, "Shooting Accuracy"].iloc[0]}')
```

# Drinks

## Read and save data from datasets/drinks.txt

```python
FILEPATH = '../data/drinks.csv'
```

## Which continent drinks more beer on average?

```python
drinks_data = pd.read_csv(FILEPATH, sep=',')
print("Drinks dataset imported from drinks.csv:\n", drinks_data)
```

## For each continent print the statistics for wine consumption.

```python
for continent in drinks_data['continent'].unique():
    continent_data = drinks_data[drinks_data['continent'] == continent]['wine_servings']
    print(f"\nWine consumption statistics for {continent}:")
    print(f"  Mean: {continent_data.mean():.2f}")
    print(f"  Min: {continent_data.min()}")
    print(f"  Max: {continent_data.max()}")
    print(f"  Std: {continent_data.std():.2f}")
```

##  Print the mean, min and max values for spirit consumption per continent.

```python
for continent in drinks_data['continent'].unique():
    continent_data = drinks_data[drinks_data['continent'] == continent]['spirit_servings']
    print(f"\nSpirit consumption statistics for {continent}:")
    print(f"  Mean: {continent_data.mean():.2f}")
    print(f"  Min: {continent_data.min()}")
    print(f"  Max: {continent_data.max()}")
```
