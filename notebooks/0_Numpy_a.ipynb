#### Stack arrays a and b vertically and then horizontally

```python
# Input
a = np.arange(10).reshape(2, -1)
b = np.repeat(1, 10).reshape(2, -1)

vertically_stacked = np.vstack((a, b))
horizontally_stacked = np.hstack((a, b))
print("Vertically Stacked:\n", vertically_stacked)
print("Horizontally Stacked:\n", horizontally_stacked)
```

#### Find positions where elements of a and b match

```python
# Input
a = np.array([1, 2, 3, 2, 3, 4, 3, 4, 5, 6])
b = np.array([7, 2, 10, 2, 7, 4, 9, 4, 9, 8])

matches = np.where(a == b)
print("Matching positions:", matches)
```

#### Compute the min-by-max (i.e. min/max) for each row for given 2d numpy array (see [apply_along_axis](https://numpy.org/doc/stable/reference/generated/numpy.apply_along_axis.html#numpy-apply-along-axis))

```python
# Input
np.random.seed(100)
a = np.random.randint(1,10, [5,3])
print(a)

min_by_max = np.min(a, axis=1) / np.max(a, axis=1)
print("Min-by-Max for each row:", min_by_max)
```

####  Find the duplicate entries in the given numpy array and mark them as True. First time occurrences should be False.

```python
# Input
np.random.seed(100)
a = np.random.randint(0, 5, 10)
print('Array: ', a)

duplicates = np.zeros_like(a, dtype=bool)
for i in range(len(a)):
    if a[i] in a[:i]:
        duplicates[i] = True
print('Duplicate entries marked as True:', duplicates)
```

#### Find the index of 5th repetition of number 1 in x
Make use of numpy.where

```python
# Input
x = np.array([1, 2, 1, 1, 3, 4, 3, 1, 1, 2, 1, 1, 2])
n = 5
repetition_indices = np.where(x == 1)[0]
if len(repetition_indices) >= n:
    index_of_nth_repetition = repetition_indices[n - 1]
    print(f"Index of {n}th repetition of number 1:", index_of_nth_repetition)
else:
    print(f"Number 1 does not repeat {n} times in the array.")
```

#### Compute the moving average of window size 3, for the given 1D array
Make use of numpy.convolve

```python
# Input
arr = np.array([8, 8, 3, 7, 7, 0, 4, 2, 5, 2])

window_size = 3
moving_average = np.convolve(arr, np.ones(window_size)/window_size, mode='valid')
print("Moving average with window size 3:", moving_average)
```

```python
# Input filepath
FILEPATH = '../data/iris.txt'
iris_data = np.genfromtxt(FILEPATH, delimiter=',', dtype=str)
print("Iris dataset imported from iris.txt:\n", iris_data)
```

#### Print the unique labels of the dataset

```python
unique_labels = np.unique(iris_data[:, -1])
print("Unique labels in the dataset:", unique_labels)
```

#### Find how many samples (i.e. rows) we have for each unique label

```python
for label in unique_labels:
    count = np.sum(iris_data[:, -1] == label)
    print(f"Number of samples for label '{label}': {count}")
```

#### Find the mean, median, standard deviation of iris's sepallength (1st column)
See np.genfromtxt, load only the 1st column with the appropiate data type

```python
sepallength = iris_data[:, 0].astype(float)
mean_sepallength = np.mean(sepallength)
median_sepallength = np.median(sepallength)
std_sepallength = np.std(sepallength)
print(f"Mean of sepallength: {mean_sepallength}")
print(f"Median of sepallength: {median_sepallength}")
print(f"Standard deviation of sepallength: {std_sepallength}")
```

####  Find the mean, median, standard deviation of iris's sepallength (1st column) for each unique labels

```python
for label in unique_labels:
    sepallength_label = iris_data[iris_data[:, -1] == label, 0].astype(float)
    mean_label = np.mean(sepallength_label)
    median_label = np.median(sepallength_label)
    std_label = np.std(sepallength_label)
    print(f"Label: {label}")
    print(f"  Mean of sepallength: {mean_label}")
    print(f"  Median of sepallength: {median_label}")
    print(f"  Standard deviation of sepallength: {std_label}")
```

####  Create a new column for volume in iris, where volume is (pi x petallength x sepal_length^2)/3

```python
petallength = iris_data[:, 2].astype(float)
sepal_length = iris_data[:, 0].astype(float)
volume = (np.pi * petallength * sepal_length**2) / 3
iris_data_with_volume = np.hstack((iris_data, volume.reshape(-1, 1)))
print("Iris dataset with new volume column:\n", iris_data_with_volume)
```

#### Randomly sample iris's species such that setosa is twice the number of versicolor and virginica
See [numpy.random.choice](https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html#numpy-random-choice)

```python
sampled_species = np.random.choice(unique_labels, size=60, p=[0.5, 0.25, 0.25])
print("Sampled species:", sampled_species)
print("\nValue counts:")
unique_sampled, counts = np.unique(sampled_species, return_counts=True)
for label, count in zip(unique_sampled, counts):
    print(f"{label}: {count}")
```

#### Find the most frequent value of petal length (3rd column) in iris dataset
